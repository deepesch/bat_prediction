```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#URL to be scraped
url_to_scrape = 'https://batgrowth.com/'
#Load html's plain data into a variable
plain_html_text = requests.get(url_to_scrape)
#parse the data
soup = BeautifulSoup(plain_html_text.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
import json
rawh = soup.find_all('h3')
```

```python
import json
rawJ = soup.find_all('script')
for i in range(4,9,1):
    J = str(rawJ[i])
    J1 = J.split('var data = ')
    J2 = J1[1].rsplit('.')[0]
    J3 = json.loads(J2)
    J4 = pd.DataFrame(J3)
    J4.Created_at = pd.to_datetime(J4['Created_at'],unit='ms')
    J4.to_csv("../data/raw/"+str(i)+'.csv',index=False)
```
